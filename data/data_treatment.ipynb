```python
import pandas as pd
import pickle

contracts_df = pd.read_csv('./contracts.csv', low_memory=False)
contracts_df.head(3)
```

```python
cols = {}
for col in contracts_df:
    cols[col] = col.upper().replace(' ', '_')

print(cols)
```

```python
contracts_df.rename(columns=cols, inplace=True)
contracts_df.head(1)
```

```python
contracts_df.isnull().sum()
```

```python
print(contracts_df['ZIP'].value_counts())
```

```python
cols = [
    'REVISION_NUMBER',
    'SPECIFICATION_NUMBER',
    'VENDOR_NAME',
    'VENDOR_ID',
    'ADDRESS_1',
    'ADDRESS_2',
    'PROCUREMENT_TYPE'
]
contracts_df.drop(columns=cols, inplace=True)
contracts_df.head(1)
```

```python
print(contracts_df['START_DATE'][1])
print(type(contracts_df['START_DATE'][1]))

contracts_df['START_DATE'] = pd.to_datetime(contracts_df['START_DATE'], errors='coerce').dt.strftime('%d/%m/%Y')
contracts_df['START_DATE'] = pd.to_datetime(contracts_df['START_DATE'], errors='coerce')
print('-------------MODIFIED--------------')
print(contracts_df['START_DATE'][1])
print(type(contracts_df['START_DATE'][1]))
```

```python
contracts_df['END_DATE'] = pd.to_datetime(contracts_df['END_DATE'], errors='coerce').dt.strftime('%d/%m/%Y')
contracts_df['END_DATE'] = pd.to_datetime(contracts_df['END_DATE'], errors='coerce', dayfirst=True)
print(type(contracts_df['END_DATE'][1]))

contracts_df.head(1)
```

```python
with open('contracts.pkl', mode='wb') as f:
    pickle.dump(contracts_df, f)
```
